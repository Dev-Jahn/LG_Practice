```
# Problem 3

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from numpy import genfromtxt


class MultivariateLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(3, 1)

    def forward(self, x):
        return self.linear(x)


my_data = genfromtxt('score.csv', delimiter=',')
x_train = torch.FloatTensor(my_data[:, :-1])
y_train = torch.FloatTensor(my_data[:, -1:])
x_test = torch.FloatTensor([70, 80, 90])

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```
# 모델 초기화
model = MultivariateLinearRegressionModel()
# optimizer 설정
optimizer = optim.SGD(model.parameters(), lr=1e-5)

nb_epochs = 500
for epoch in range(nb_epochs + 1):    # H(x) 계산
    prediction = model(x_train)
    cost = F.mse_loss(prediction, y_train)  

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()  # gradient 계산
    optimizer.step()  # update

    # 20번마다 로그 출력
    if epoch % 20 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()		#list(model.parameters())[0].data[0][0]
        ))

print("Answer: ", model(x_test).data[0])
```
