```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```
x_train = torch.FloatTensor([[1], 
                             [2], 
                             [3]])
y_train = torch.FloatTensor([[2], 
                             [3], 
                             [4]])
# y = W * x + b
# 모델 초기화
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

print(W)
print(b)
```

```
# optimizer 설정
optimizer = optim.SGD([W, b], lr=0.01)

nb_epochs = 1000
```

```
for epoch in range(nb_epochs + 1):

    # H(x) 계산
    hypothesis = x_train * W + b

    # cost 계산
    cost = torch.mean((hypothesis - y_train) ** 2)  # Error, Cost, Loss Function

    # cost로 H(x) 개선
    optimizer.zero_grad()  # 1 : gradient 초기화

    cost.backward()  # 2 : gradient 계산

    if epoch % 100 == 0:
        print("W:", W, "W의 Gradient:", W.grad) 
        print("b:", b, "b의 Gradient:", b.grad)
    
    optimizer.step()  # 3 : gradient 빼서 update
    
    if epoch % 100 == 0:
        print("W:", W, "W의 Gradient:", W.grad) 
        print("b:", b, "b의 Gradient:", b.grad)
   
    # 100번마다 로그 출력
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f}, Cost: {:.6f}'.format(
            epoch, nb_epochs, W.item(), b.item(), cost.item()
        ))
```
