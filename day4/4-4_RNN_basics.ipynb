```python
import torch
import numpy as np
```

```python
# declare dimension
input_size = 4
hidden_size = 2
```

```python
# singleton example
# shape : (1, 1, 4)
# input_data_np = np.array([[[1, 0, 0, 0]]])

# sequential example
# shape : (3, 5, 4)
# (batch size, seq length, input_size)
h = [1, 0, 0, 0]
e = [0, 1, 0, 0]
l = [0, 0, 1, 0]
o = [0, 0, 0, 1]
input_data_np = np.array([[h, e, l, l, o], 
                          [e, o, l, l, l], 
                          [l, l, e, e, l]], dtype=np.float32)  # (3, 5, 4)
```

```python
# transform as torch tensor
input_data = torch.Tensor(input_data_np)
```

```python
# declare RNN
rnn = torch.nn.RNN(input_size, hidden_size, batch_first=True)  # 처음에 batch_size or seq_length
```

```python
# check output
outputs, _status = rnn(input_data)
# inputs: input, h_0 = 0 (num_layers*num_directions, batch, hidden_size)

print(input_data.size())

print(outputs.size()) # outputs.shape

print(torch.transpose(_status, 1, 0).size())

# print(outputs[:, -1])
# print(_status)
```

```python
lstm = torch.nn.LSTM(input_size, hidden_size, batch_first=True)
outputs, _status = lstm(input_data)
print(outputs.size())
print(torch.transpose(_status[0], 1, 0).size())

# print(outputs[:, -1])
# print(_status[0][0])
```

```python
gru = torch.nn.GRU(input_size, hidden_size, batch_first=True)
outputs, _status = gru(input_data)

print(outputs.size())
print(torch.transpose(_status, 1, 0).size())

# print(outputs[:, -1])
# print(_status[0])
```

```python
# declare Stacked LSTM
stackedlstm = torch.nn.LSTM(input_size, hidden_size, num_layers=6, batch_first=True)  # 처음에 batch_size or seq_length
outputs, _status = stackedlstm(input_data)

print(outputs.size())

print(torch.transpose(_status[0], 1, 0).size())

# print(outputs[:, -1])
# print(_status[0][5])
```

```python
# declare Bidirectional LSTM
bidirectionLSTM = torch.nn.LSTM(input_size, hidden_size, bidirectional=True, batch_first=True)  # 처음에 batch_size or seq_length
outputs, _status = bidirectionLSTM(input_data)

print(outputs.size())

print(torch.transpose(_status[0], 1, 0).size())
```
